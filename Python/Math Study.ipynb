## Linear Algebra with Python

Two objects are heading in the same direction at different speeds and times of departure. If Object 1 were traveling at 150km/hr and leaving at t = 0, and Object 2 was moving at 180 km/hr leaving at t = 5, how long would it take the two objects to meet? (Where t is in mins)

To solve this problem, we could first convert their speeds into km/min

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
speed_obj1 = 150 / 60 
speed_obj2 = 180 / 60
```

Ignoring acceleration and other potential effects on speed, we can assume the equation for the distance traveled by Object 1 is d = s * t (where speed = s and time = t). With a five minute difference in departure, the equation for Object 2 would thus be d = s * (t - 5).

```python
t = np.linspace(0, 40, 1000)
```

```python
d_1 = speed_obj1 * t
```

```python
d_2 = speed_obj2 * (t - 5)
```

This problem can be solved graphically as well as algebraically. Python makes graphing such problems fairly easy, however, it is quite difficult to create a graph that clearly shows important information without first solving the problem algebraically.

By setting the two equations equal to one another, we can solve for the variable of distance.

2.5t = 3.0(t - 5)

-0.5t = -15

t = 30

We now know the intercept point of the two objects, and thus by inputting the value for t we can find the distance traveled.

d_1 = 2.5 * (30)

d_1 = 75

We can now use our knowledge to create a graph that is appropriately scaled and thus better able to communicate its information. Withoout linear algebra, we would have spent as much playing with the variables as we would have solving the problem algebraically

```python
fig, ax = plt.subplots()
plt.title('Intercept of Object 1')
plt.xlabel('time (in minutes)')
plt.ylabel('distance (in km)')
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(t, d_1, c='green')
ax.plot(t, d_2, c='brown')
plt.axvline(x=30, color='purple', linestyle='--')
_ = plt.axhline(y=75, color='purple', linestyle='--')
```

For a more practical application, we can imagine that we are a part of the mission to terraform Mars. We are tasked with developing solar panels for the mission and our first prototype produces 1 kJ of energy per day. Our prototype is launched and begins producing power on April 1st. However, shortly after launch we develop a second prototype that produces 4 kJ of energy per day. The second prototype arrives on the planet and begins producing energy on May 1st.

Mission control requests the following information:

    1. What day will the Mk.2 prototype produce as much power as the Mk.1?
    
    2. What is the total energy produced by the panels at this point?
    
    3. What would the solution be to the first two problems if the Mk.2 produced 1 kJ/day?
    

Using linear algebra, we can solve this problems and then create visualizations of the data that will best convey the information to mission control.

With 1kJ/day, the Mk.1 would have the following equation (where t = days and e is energy):
    
    e = 1t
    
The Mk.2, with a thirty day latency, would have the following equation:
    
    e = 4(t - 30)
    
By setting the two equations equal to one another, we solve for the point at which the energy output is the same.

    1t = 4(t - 30)
    
    t = 4t - 120
    
    -3t = -120
    
    t = 40
    
Thus at day 40 the two panels would have produced the same amount of energy.

The total energy produced can be reasonably deduced to be 80 k/J.

```python
# Lets code the above equations
t2 = np.linspace(0, 100, 1000)
mk1 = 1 * t2
mk2 = 4 * (t2 - 30)
```

```python
fig, ax = plt.subplots()
plt.title('Power Output of Mk.1 & Mk.2')
plt.xlabel('Time (in days)')
plt.ylabel('Energ(in kJ)')
ax.set_xlim([0, 60])
ax.set_ylim([0, 100])
ax.plot(t2, mk1, c='green')
ax.plot(t2, mk2, c='brown')
plt.axvline(x=40, color='purple', linestyle='--')
_ = plt.axhline(y=40, color='purple', linestyle='--')
```

To answer the final question posed by mission control, lets use our graphing skills to examine what happens.

```python
new_mk2 = 1 * (t2 -30)
```

```python
fig, ax = plt.subplots()
plt.title('Power Output of Mk.1 & Mk.2')
plt.xlabel('Time (in days)')
plt.ylabel('Energ(in kJ)')
ax.set_xlim([0, 100])
ax.set_ylim([0, 100])
ax.plot(t2, mk1, c='green')
ax.plot(t2, new_mk2, c='brown')
plt.axvline(x=40, color='purple', linestyle='--')
_ = plt.axhline(y=40, color='purple', linestyle='--')
```

Because the slope of each line is the same, the lines are parallel, meaning that there is no solution.
