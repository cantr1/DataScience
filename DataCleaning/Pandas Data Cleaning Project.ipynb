Data, in its raw form, often comes with various imperfections—missing values, inconsistencies, or anomalies—that can affect the outcome and reliability of analytics processes. Born from the necessity to handle large datasets with ease and flexibility, Pandas has become the de facto tool for data manipulation in Python. Its data structures provide an intuitive interface to slice and dice data, reshape datasets, and perform aggregations, among other tasks.

### In this project, I was given a dataset of potential clients on a call list that will later be used by our sales team for contacting these individuals. Management would like this dataset to be cleaned so that the sales team can more easily and efficiently make use of the data.

# First Steps - Import the Library and Create a DataFrame

```python
# Import Library
import pandas as pd
```

```python
# Pandas will read the Excel file and assign it to the variable 'df'
df = pd.read_excel(r"C:\Users\unbou\Desktop\PresentationProjects\DataCleaning\PandasCleaning\Customer Call List.xlsx")
```

```python
# Printing the dataframe for a quick overview
df
```

# Initial Cleaning

```python
# The first step to cleaning most datasets to remove duplicates
# This function creates a new dataset with no duplicates
# It is important not to forget to set our dataframe equal to this code
df = df.drop_duplicates()
```

```python
# Next it would be wise to remove data not useful to our sales team
# We can utilize the .drop() function to select columns in our dataset
# for removal
df = df.drop(columns = "Not_Useful_Column")
```

# Advanced Cleaning

### Last Name

```python
# Looking at the dataset, it appears the column "Last_Name" has some issues
df["Last_Name"]
```

```python
# We can use the .strip() function to remove these inconsistencies
# An important deviation from the previous workflow is that we have to set
# the column equal to this code to avoid errasing our entire dataframe
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df["Last_Name"]
```

### Phone Numbers

```python
# We want each phone number to follow the same format
# To do so we can remove any non-numeric symbols so that each phone number
# is simply a string of numbers we can later format
df['Phone_Number'] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('[^\d]', '', regex=True)
```

```python
# We must first convert each number into a string
df["Phone_Number"] = df["Phone_Number"].apply(lambda x : str(x))
```

```python
# We can now use a lambda to format each number into the same format
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:10])
```

```python
# Our final step is to remove the 'nan' and '--'
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--', '')
df["Phone_Number"] = df["Phone_Number"].str.replace('--', '')
```

```python
# The 'Phone_Number' column is now clean!
df
```

## Address 

```python
# Sales would also like the address information split into multiple columns
# to get a better picture of where each client lives at a glance
# We c
df[["Street", "State", "Area Code"]] = df["Address"].str.split(',',n=2, expand=True)
df
```

## Paying Customer & Do Not Contact

```python
# For consistency, the 'PayingCustomer' column should be converted so that
# all values follow a similar format
df["Paying Customer"] = df["Paying Customer"].str.replace("Yes", "Y")
df["Paying Customer"] = df["Paying Customer"].str.replace("No", "N")

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes", "Y")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No", "N")

df
```

# Filtering the Data

### It is important that the data be filtered such that the list that gets handed to the sales team only contains clients who they are both able to contact and those that want to be contacted.

## Remove Empty Values

```python
# Removes values that are empty
df = df.fillna('')
```

## Remove Clients that are DNC

```python
# Create a for loop that identifies clients who do not want to be contacted
# and removes them from the list.
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace = True)
        
df
```

```python
# Remove clients that do not have a number listed
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace = True)
        
df
```

```python
# Reset the index to get a better idea of the length of the list
# and make the list more visually appealing
df = df.reset_index(drop=True)
```
