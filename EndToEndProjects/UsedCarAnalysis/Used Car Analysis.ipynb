# Used Car Data

## In this project I will be using the pandas library to further clean the data before creating visualizations in Tableau

```python
# Import Libraries
import pandas as pd
```

```python
# Create a DataFrame
df = pd.read_csv("train_cleaned.csv")
```

```python
# Show data
df
```

```python
# Drop null values and remove duplicates
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
```

```python
# Remove values of 0
df = df[(df != 0).all(axis=1)]
```

```python
# The cleaned data frame
df
```

```python
# Exports the newly created file into an Excel format
df.to_excel('fully_cleaned_data.xlsx', index=False)
```
